## Questões a serem resolvidas

### [Questão 1] Mostre o nome de todos os restaurantes sem repetir. Separadamente, mostre também todos os tipos distintos de cozinhas.

````python
pprint(  db.restaurants.distinct("name")  )
````

````python
pprint ( db.restaurants.distinct("cuisine") );
````

### [Questão 2] Mostre o nome e o tipo de cozinha de 5 restaurantes quaisquer sem exibir o campo "_id"

````python
result = db.restaurants.find({}, {"name":1, "cuisine": 1, "_id": 0}).limit(5)
pprint( list(result) );
````

### [Questão 3] Mostre 5 restaurantes no bairro "Bronx"

````python
result = db.restaurants.find({ 'borough': 'Brooklyn' }, {"name": 1, "borough": 1, "_id": 0}).limit(5)
pprint( list(result) );
````

### [Questão 4] Altere o tipo de cozinha do restaurante "Kosher Island" para "Sea food"

````python
db.restaurants.update_one({'name': "Kosher Island"}, { "$set": { "cuisine": "Sea food" } })
result = db.restaurants.find_one({'name': "Kosher Island"})
pprint( result )
````

### [Questão 5] Mostre e remova o(s) restaurante(s) com zipcode "10314"

````python
results = db.restaurants.find({ "address.zipcode": "10314" });
pprint ( list(results) )

db.restaurants.delete_many({ "address.zipcode": "10314" })

results = db.restaurants.find({ "address.zipcode": "10314" });

pprint( list(results))
````

### [Questão 6] Adicione um campo status ao restaurante cujo nome é "Riviera Caterer". Status="demolido".

````python
result = db.restaurants.find_one({"name": "Riviera Caterer"})
pprint( result)

db.restaurants.update_one({"_id": result.get("_id")}, {"$set":  { "status": "demolido" } })

result = db.restaurants.find_one({"name": "Riviera Caterer"})

pprint(result)
````
